## Univariate Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame

num_cols = df.select_dtypes(include=[int, float]).columns.tolist()
num_cols.pop()

for col in num_cols:
    plt.figure()
    plt.title(f'Distribution of {col}')
    plt.hist(df[col], bins=30)  # Adjust bins as needed
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

# Box Plots (numeric)

```python
for c in num_cols:
    plt.figure()
    plt.boxplot(df[c].dropna(), vert=True)
    plt.title(f"Box plot: {c}")
    plt.ylabel(c)
    plt.show()
```

## 3) Outlier Detection

Z-Score Method (|z| > 3)

```python
z = np.abs(stats.zscore(df[num_cols], nan_policy="omit"))
z_outliers_mask = (z > 3).any(axis=1)
print("Z-score outlier rows:", int(z_outliers_mask.sum()))
```

```python
# Keep only rows where all |Z| <= 3 (no outliers)
df_clean = df[(z < 3).all(axis=1)]

print("After removing Z-score outliers:", df_clean.shape)

# 5 Save cleaned dataset
df_clean.to_csv("../data/processed/togo_clean.csv", index=False)

print("Cleaned dataset saved to: data/processed/togo_clean.csv")
```

## Correlation Matrix & Heatmap (numeric)

```python
corr = df[num_cols].corr()
display(corr)

plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

##  Correlation Matrix  Interpretation
Interpretation of the correlation matrix you showed (variables: GHI, DNI, DHI, ModA, ModB, Tamb, RH, WS, WSgust, WSstdev, WD, WDstdev, BP, Cleaning, Precipitation, TModA, TModB).

Main patterns
 Strong positive cluster of solar/irradiance and module-temperature variables:
   GHI, DNI, DHI, ModA, ModB, TModA and TModB are highly positively correlated with each other (many correlations  0.81.0). In plain terms: when one irradiance or modeled-module-temperature variable rises, the others almost always rise too.
 Ambient temperature (Tamb) is moderately positively correlated with irradiance variables ( +0.5). So higher solar input tends to coincide with higher air temperature.
 Relative humidity (RH) is negatively correlated with irradiance and temperature (roughly -0.2 to -0.4). That means clearer, sunnier conditions (higher GHI/DNI/Tamb) tend to have lower RH in this dataset.
 Wind variables form their own correlated group:
   WS (wind speed), WSgust and WSstdev are very strongly correlated with each other (WSWSgust  0.98; WSstdev also strongly related). These three move together.
   Wind metrics show modest positive correlations with irradiance (WS vs GHI  +0.4), i.e., wind is somewhat higher on sunnier days in your data.
 Wind direction (WD) and its variability (WDstdev) show weak/variable correlations with other variables (not major drivers here).
 Barometric pressure (BP) shows a noticeable negative correlation with ambient temperature (BP vs Tamb  -0.5), indicating lower pressure tends to coincide with higher temperatures in your data.
 Cleaning and Precipitation have very small correlations with most variables (near zero), suggesting they are largely independent in this dataset or have little simultaneous variation at the measurement scale used.

Practical implications
 Multicollinearity risk: Many irradiance-related variables are almost redundant (very high pairwise correlations). If you plan to build predictive models (e.g., for power output), include only one or a small orthogonal set of these (or use PCA / regularized regression). Using several highly correlated predictors will inflate variance and complicate coefficient interpretation.
 Variable selection guidance:
   For solar/production modeling, pick a single irradiance measure (GHI or DNI) or the model output (ModA/ModB) and one temperature variable (Tamb or TMod). Avoid using all of them together.
   Wind features can be reduced: keep WS or WSgust (they are nearly interchangeable), plus WSstdev only if variability matters.
   Cleaning and Precipitation can likely be ignored as predictors unless you have reason to expect lagged or nonlinear effects.
 Physical insights:
   Sunny, warm, low-humidity conditions are tightly coupled in your dataset.
   Wind tends to be higher on sunnier days here (but correlation is modest).
